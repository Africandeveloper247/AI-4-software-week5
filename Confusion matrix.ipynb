```python

# Hypothetical confusion matrix values
TP = 40   # True Positives
FP = 20   # False Positives
FN = 10   # False Negatives
TN = 80   # True Negatives

# Display confusion matrix
print("Confusion Matrix:")
print(f"                Predicted No   Predicted Yes")
print(f"Actual No       {TN:>10}       {FP:>10}")
print(f"Actual Yes      {FN:>10}       {TP:>10}")

# Calculate precision and recall
precision = TP / (TP + FP)
recall = TP / (TP + FN)

print("\nPrecision:", round(precision, 2))
print("Recall:", round(recall, 2))
```
